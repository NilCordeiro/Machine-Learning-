```python
import seaborn as sns  # importando a biblioteca seaborn
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')    # o dataset tips ele nós dar informações sobre gorjetas de determindado restaurante
tips.head()
```

### Plot linear model ou lmplot

```python
sns.lmplot(x ='total_bill',y = 'tip',data = tips,hue = 'sex') # plotando o grafico de regressão
```

```python
sns.lmplot(x ='total_bill',y = 'tip',data = tips,hue = 'sex',col = 'time') # podemos tambem dividir o plot em coluna com uma variavel categorica
```

```python
sns.lmplot(x ='total_bill',y = 'tip',data = tips,hue = 'sex',row = 'time') # podemos tambem plotar o rafico pela linha escolhendo um variavel 
```

É isso turma plots de regressão
