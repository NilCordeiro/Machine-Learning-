Agora vamos discutir como usar seaborn para traçar dados categóricos.Existem alguns tipos de argumentos principais para isso:

<ul>
    <li> factorplot
    <li> boxplot
    <li> violinplot
    <li> stripplot
    <li> swarmplot
    <li> barplot
    <li> countplot
</ul>

Vamos passar por exemplos de cada um.

```python
import seaborn as sns
import numpy as np   # importando as bibliotecas necesssarias
```

```python
tips = sns.load_dataset('tips') # vamos importar o dataset tips novamente para trabalharmos com os plots categóricos
tips.head() # gerando o cabeçalho do dataset
```

## barplot

```python
sns.set(style = 'whitegrid')
sns.barplot(x = 'sex',y ='total_bill',data = tips)  # por padrão o barplot esses valores é o valor médio, podemos ver que a média da conta do sexo masculino é 21
```

## countplot

```python
sns.countplot(x = 'sex',data = tips) # plot que faz a contagem da variavel categórica, não precisa do espesificar o eixo y sendo o mesmo é a contagem
# nesse plot podemos ver que na variavel sex temos uma quantidade maior do sexo masculino 
```

## boxplot

```python
sns.boxplot(x = 'day',y = 'total_bill',data = tips)
```

## violinplot

```python
sns.violinplot(x = 'day',y = 'tip',data = tips)
```

```python
sns.violinplot(x = 'day',y = 'tip',data = tips,hue = 'sex',split = True)# poemos tambem quebrar os dados e ter melhor uma visualização, com o metodo hue para a variavel sex
```

## stripplot

```python
sns.stripplot(x = 'day',y = 'total_bill',data = tips,jitter = True,hue = 'sex',split = True)
```

## swarmplot

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex',split = True)
```

## factorplot

```python
sns.factorplot(x = 'sex',y = 'total_bill',data = tips,kind = 'bar') # é um plot bem paresido com o barplot
```
