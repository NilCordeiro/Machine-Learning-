## Plots de Distribuição:

Vamos discutir alguns graficos que nos permitem visualizar a distrobuição de dados esses plots são:

<ul>
    <li> distplot
    <li> jointplot
    <li> pairplot
</ul>

```python
import seaborn as sns # importando a biblioteca seaborn
sns.set(style = 'whitegrid') # configurando a cor dos plots
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')  # Podemos importar um dataset pronto do seaborn, nesse exemplo importarmos o dataset tips ou gorjeta
tips.head()
```

```python
tips.info()  # Gerando informações importantes sobre o dataset,podemos ver que não tem dados faltantes, tem um total de 7 colunas, 244 linhas  
```

```python
sns.distplot(tips['total_bill'],kde = False) # plotando um grafico de distribuição na coluna total da conta do dataset 
```

### Jointplot

O jointplot cria uma distribuição com duas variaveis ao mesmo tempo

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips)# podemos observar que temos uma pequena correlação entre as variaveis, onde quanto maior a conta maior a gorjeta
```

### Pairplot 

O pairplot pega todas as variaveis numericas e faz um jointplot um distribuição

```python
sns.pairplot(tips,hue = 'sex') # com o atributo hue podemos separar o sexo homens e mulheres 
```
